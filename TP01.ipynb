```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = x.set_index('abbr')
```

```python
x.plot.bar(color=['b', 'r', 'g'] , figsize=(15,15))
```

```python
tc = x.plot.bar(y='Total Cases', figsize=(15,15))
tc
```
